```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('df4.csv')

# Access and manipulate the data in the DataFrame as needed
# For example, you can print the first few rows of the DataFrame
df
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis to each document in the filtered_df
sentiment_scores = []

for index, row in df.iterrows():
    document = row['document']
    score = sia.polarity_scores(document)
    sentiment_scores.append(score)

# Add the sentiment scores as a new column to filtered_df
df['sentiment_scores'] = sentiment_scores
```

```python
sentiment_scores
```

```python

df1 = pd.DataFrame.from_dict(sentiment_scores)

df1
```

```python
def classify_sentiment(compound_score):
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'
```

```python
df1['sentiment'] = df1['compound'].apply(classify_sentiment).reset_index(drop=True)
df1
```

```python
df1.to_csv('sentiment.csv', index=False)
```

```python
df.to_csv('original_df.csv', index=False)
```

```python
final_df = pd.concat([df1["sentiment"], df], axis=1)

final_df = final_df.drop(columns = ["sentiment_scores"])
```
