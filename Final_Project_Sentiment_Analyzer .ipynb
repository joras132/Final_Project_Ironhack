```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('output.csv')

# Access and manipulate the data in the DataFrame as needed
# For example, you can print the first few rows of the DataFrame
print(df.head())
```

```python
filtered_df = df[df['topic'] == 1]
filtered_df
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download('vader_lexicon')

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Apply sentiment analysis to each document in the filtered_df
sentiment_scores = []

for index, row in filtered_df.iterrows():
    document = row['document']
    score = sia.polarity_scores(document)
    sentiment_scores.append(score)

# Add the sentiment scores as a new column to filtered_df
filtered_df['sentiment_scores'] = sentiment_scores
```
